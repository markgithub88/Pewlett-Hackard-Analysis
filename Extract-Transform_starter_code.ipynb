```python
backerdf['backer_id'] = backerdf['backer_info'].str.extract('([a-zA-Z]{2}\d{3})')

backerdf['cf_id'] = backerdf['backer_info'].str.extract('(\s\d{1,4})')
backerdf.head()
```

```python
backerdf['email'] = backerdf['backer_info'].str.extract('([a-zA-Z]+@[a-zA-Z]+\.com)')
backerdf.head()
```
